```python
# 22/mayo/2026

# Programador:
# Lenguaje de programación, experiencia,
# tipo de proyecto, dificultad y análisis IA.

# ======================================================
# 💻 CBTis 128
# 📚 Elabora Proyectos de Programación Lógica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Manuel Nava
# 👨‍🎓 Alumno: Axel Guzman NL = 27
# 💻 Proyecto: Lista de Diccionarios Programador IA
# ======================================================

# =========================
# 💻 Lista vacía
# =========================
programadores = []

# =========================
# 📚 Diccionario vacío
# =========================
losprogramadores = {}

# =========================
# 📌 Menú principal
# =========================
while True:

    print("\n=========== MENÚ PROGRAMADOR 💻 ===========")
    print("1. Agregar programador")
    print("2. Mostrar programadores")
    print("3. Buscar programador")
    print("4. Analizar programador")
    print("5. Salir")

    opcion = int(input("Selecciona una opción: "))

    # ==================================================
    # ➕ AGREGAR PROGRAMADOR
    # ==================================================

    if opcion == 1:

        clave = input("Ingrese clave del programador: ")
        nombre = input("Ingrese nombre del programador: ")
        lenguaje = input("Ingrese lenguaje de programación: ")
        experiencia = int(input("Ingrese años de experiencia: "))
        proyecto = input("Ingrese tipo de proyecto: ")
        dificultad = input("Ingrese dificultad (bajo, medio, alto): ")

        # Diccionario
        losprogramadores = {

            "Nombre": nombre,
            "Lenguaje": lenguaje,
            "Experiencia": experiencia,
            "Proyecto": proyecto,
            "Dificultad": dificultad
        }

        # Lista de diccionarios
        programadores.append({

            "Clave": clave,
            "Nombre": nombre,
            "Lenguaje": lenguaje,
            "Experiencia": experiencia,
            "Proyecto": proyecto,
            "Dificultad": dificultad
        })

        print("\n✅ Programador agregado correctamente")

    # ==================================================
    # 📋 MOSTRAR PROGRAMADORES
    # ==================================================

    elif opcion == 2:

        print("\n===================================================================================================================")
        print(f"{'CLAVE':<12}{'NOMBRE':<20}{'LENGUAJE':<15}{'EXPERIENCIA':<15}{'PROYECTO':<20}{'DIFICULTAD':<15}")
        print("===================================================================================================================")

        i = 0

        while i < len(programadores):

            print(f"{programadores[i]['Clave']:<12}"
                  f"{programadores[i]['Nombre']:<20}"
                  f"{programadores[i]['Lenguaje']:<15}"
                  f"{programadores[i]['Experiencia']:<15}"
                  f"{programadores[i]['Proyecto']:<20}"
                  f"{programadores[i]['Dificultad']:<15}")

            i += 1

        print("===================================================================================================================")

    # ==================================================
    # 🔍 BUSCAR PROGRAMADOR
    # ==================================================

    elif opcion == 3:

        buscar = input("Ingrese clave del programador: ")

        i = 0
        encontrado = False

        while i < len(programadores):

            if programadores[i]["Clave"] == buscar:

                print("\n=========== PROGRAMADOR ENCONTRADO ===========")
                print("Nombre:", programadores[i]["Nombre"])
                print("Lenguaje:", programadores[i]["Lenguaje"])
                print("Experiencia:", programadores[i]["Experiencia"])
                print("Proyecto:", programadores[i]["Proyecto"])
                print("Dificultad:", programadores[i]["Dificultad"])

                encontrado = True

            i += 1

        if encontrado == False:
            print("❌ Programador no encontrado")

    # ==================================================
    # 🤖 ANALIZAR PROGRAMADOR
    # ==================================================

    elif opcion == 4:

        experiencia = int(input("Ingrese experiencia: "))
        lenguaje = input("Ingrese lenguaje: ")
        dificultad = input("Ingrese dificultad: ")

        print("\n=========== ANALISIS IA ===========")

        # IF
        if experiencia >= 5:
            print("💻 Programador avanzado")

        # IF ELSE
        if dificultad == "alto":
            print("⚠️ Proyecto complejo")
        else:
            print("✅ Proyecto manejable")

        # IF ELIF ELSE
        if lenguaje == "python":
            print("🤖 IA recomienda Machine Learning")

        elif lenguaje == "java":
            print("☕ IA recomienda aplicaciones empresariales")

        elif lenguaje == "javascript":
            print("🌐 IA recomienda desarrollo web")

        else:
            print("💻 Lenguaje válido para programación general")

        # Operadores lógicos
        if experiencia >= 3 and dificultad == "alto":
            print("🔥 IA detecta programador apto para proyectos complejos")

        elif experiencia < 3 or dificultad == "medio":
            print("📚 IA recomienda seguir practicando programación")

    # ==================================================
    # ❌ SALIR
    # ==================================================

    elif opcion == 5:

        print("\n💻 Programa finalizado")
        break

    # ==================================================
    # ⚠️ OPCIÓN INVÁLIDA
    # ==================================================

    else:

        print("❌ Opción no válida")
```
